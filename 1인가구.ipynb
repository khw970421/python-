```python
import csv
import matplotlib.pyplot as plt

f= open('1인.csv')
data=csv.reader(f)
a=[]
b=[]
c=[]
d=[]
ap=[]
bp=[]
cp=[]
dp=[]
next(data)
next(data)
next(data)
for row in data:
    for i in range(12):
        if i==3:
            a.append(float(row[i]))
            ap.append(row[0])
        if i==6:
            b.append(float(row[i]))
            bp.append(row[0])
        if i==9:
            c.append(float(row[i]))
            cp.append(row[0])
        if i==12:
            d.append(float(row[i]))
            dp.append(row[0])
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('연도별 도시의 1인가구 비율')
plt.xticks(rotation=90)
plt.yticks()
plt.plot(ap,a)
plt.plot(ap,b)
plt.plot(ap,c)

plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f= open('1인.csv')
data=csv.reader(f)
a=[]
b=[]
c=[]
d=[]
place=[]

next(data)
next(data)
next(data)
for row in data:
    for i in range(13):
        if i==3:
            a.append(float(row[i]))
            place.append(row[0])
        if i==6:
            b.append(float(row[i]))
        if i==9:
            c.append(float(row[i]))
        if i==12:
            d.append(float(row[i]))
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('연도별 도시의 1인가구 비율')
plt.xticks(rotation=90)
plt.yticks()
plt.plot(place,a,label='2015')
plt.plot(place,b,label='2016')
plt.plot(place,c,label='2017')
plt.plot(place,d,label='2018')

plt.legend()
plt.show()
```
