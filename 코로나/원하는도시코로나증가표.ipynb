```python
import csv
import matplotlib.pyplot as plt
f=open('covidcity.csv',encoding='UTF8')
result=[]

data=csv.reader(f)
date=[]
for row in data:
    if '구분' in row[0]:
        for i in row[1:]:
            date.append(str(i)) #유지보수 간편(계속 데이터가 늘어도 date를 생성하기때문)
f.close()
                                #전체를 다 for row in data로 썼으므로 다시 닫고 다시 열고 원하는 부분 실행
f=open('covidcity.csv',encoding='UTF8')
data=csv.reader(f)
next(data)
num=0                           #num을 지정해줌으로써 date가 늘어나 num도 증가하기 때문에 range(num)에 숫자를 매번바꿀필요가없음
name=input('궁금한 도시를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in row[1:]:
            result.append(int(i))
            num=num+1
plt.rc('font',family='Malgun Gothic')
plt.plot(range(num),result)
plt.xticks(range(num),date,rotation=90)
plt.show()
```
