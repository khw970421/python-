```python
import csv
import matplotlib.pyplot as plt

f=open('covidcity.csv',encoding='UTF8')
data=csv.reader(f)
date=[]
dnum=0
for row in data:
    if '구분' in row[0]:
         for i in row[1:]:
                dnum+=1
                date.append(str(i))                    #date는 각각의 날짜를 증가 dnum은 날짜의 개수
                 
f.close

f=open('covidcity.csv',encoding='UTF8')
data=csv.reader(f)                          
dcovidnum=[0]*dnum                                     #날짜의 개수만큼 날짜별 코로나 인구수 리스트를 만들어낸다
next(data)

for row in data:
    for i in range(dnum):
        dcovidnum[i]+=int(row[i+1])                    #날짜별 코로나 인구수 리스트에 각각의 날짜에 따른 인구수 증가를 한다
        
plt.rc('font',family='Malgun Gothic')
plt.title('코로나 날짜별 증가')
plt.plot(date,dcovidnum)                                #가로는 코로나 날짜  세로는 날짜별 코로나 인구수 증가를 처리
plt.xticks(rotation=90)                                 #가로에 있는 날짜를 글자 겹침방지를 위해 90도 회전
plt.show()
```
