```python
import csv
import matplotlib.pyplot as plt

f = open('sub.csv')
data = csv.reader(f)
next(data)
ss=input('원하는 승차비율들을 알기위한 역의 이름을 입력하시오: ')
result=[0]*4
for row in data:
        if ss in row[3]:
            for i in range(4,8):
                result[i-4]=int(row[i])
                results=row[3]
      
print(result[0],result[1],result[2],result[3])
c=['#14CCC0','#389993','#FF1C6A','#CC14AF']
label=['유임승차','유임하차','무임승차','무임하차']            
plt.rc('font',family='Malgun Gothic')
plt.title(results)
plt.pie(result[0:4],colors=c,labels=label,autopct='%1.f%%')
plt.show()
        
```
